```python
import os
import time
import pandas as pd
import re
import json
```

```python
cities_data = None
with open('../cidades/cities.json') as f:
    cities_data = json.loads(f.read())
```

```python
df = pd.read_csv('../processed/targets.csv')
df['UF'] = None
df['Comarca'] = None
```

```python
def find_state(text):
    for state in cities_data:
        if state in text:
            return state

    return None


def find_city(state, text):
    for city in cities_data[state]:
        if city in text:
            return city
    return None
    
```

```python
# time1 = time.time()

state_miss = 0
comarca_miss = 0
total = 0

# for filename in os.listdir('data'):
for index, filename in enumerate(df.Files.values):
    if os.path.isfile('../data/{}'.format(filename)):
        with open('../data/{}'.format(filename), encoding="windows-1252") as f:
            text = f.read()
            states = re.findall(r'estado [\w\t\r\n]*\s*[^\n\r]*', text, re.IGNORECASE)
            cities = re.findall(r'comarca\s?\:?[\w\t\r\n\:]*\s*[^\n\r]*', text, re.IGNORECASE)
            if len(states):
                text = ''.join(states[0])
                uf = find_state(text.lower().strip())
                df.UF.set_value(index, uf)
                if len(cities) and uf:
                    text = ''.join(cities[0])
                    comarca = find_city(uf.lower().strip(), text.lower().strip())
                    df.Comarca.set_value(index, comarca)
                else:
                    comarca_miss += 1
            else:
                print(filename)
                state_miss += 1
                comarca_miss += 1

            total += 1

            print('{}, {:.2f}, {:.2f}'.format(total, 100 * state_miss/total, 100 * comarca_miss/total))
                
df.to_csv('../processed/proc.csv')
# time2 = time.time()
# print('{:.3f} min'.format((time2 - time1) / 60))
```
