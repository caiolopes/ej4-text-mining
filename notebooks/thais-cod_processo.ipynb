```python
import re as regex
from glob import glob
from os import path
```

```python
DIR = "../data/"
cods = []
for file in glob(path.join(DIR, '*.txt')):
    cods.append(regex.sub(r"\_.\.txt","",file.replace(DIR,"")))
```

```python
cods = set(cods)
```
